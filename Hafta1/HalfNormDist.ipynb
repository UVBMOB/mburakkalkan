```python
from scipy.stats import halfnorm
import seaborn as sbn

class HalfNormDist:
    
    def generate_data(self, size):
        self.data = halfnorm.rvs(size = size)
        self.mean = self.data.mean()
        self.std = self.data.std()
        
    def plot(self):
        chart = sbn.distplot(self.data)
        chart.set(xlabel="Values", ylabel="Frequency")
    
    def pdf(self, x):
        return halfnorm(self.mean, self.std).pdf(x)
        
    def cdf(self, x):
        return halfnorm(self.mean, self.std).cdf(x)

test = HalfNormDist()
test.generate_data(size=1000)
test.plot()
print('PDF: %.3f' % test.pdf(1))
print('CDF: %.3f' % test.cdf(1))
```
